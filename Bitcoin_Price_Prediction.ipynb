```
d1 = d1.fillna(d1.mean())
d1 = d1.drop(['Date'],axis=1)
```

```
d1.head()
```

```
d2 = d1[["btc_market_cap","btc_n_transactions","btc_miners_revenue","btc_cost_per_transaction","btc_difficulty","btc_hash_rate","btc_cost_per_transaction_percent","btc_market_price"]]
```

```
d2.head()
```

```
x = d2.iloc[:,:-1].values
y = d2.iloc[:,-1].values
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y)
```

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
reg = lr.fit(X_train,Y_train)
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,reg.predict(X_test))
```
